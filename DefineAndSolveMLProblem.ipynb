```python
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
```

<b>Task:</b> Use the rest of this notebook to carry out your project plan.

```python
features = ['Happiness', 'Support', 'Life']

X = df_scaled[features]
y = df_scaled['LogGDP']

print(X)
print(y)
```

```python
#split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1234)
```

```python
#create linear regression model
model = LinearRegression()

#fit model to training data 
model.fit(X_train, y_train)

#make predictions on test data
prediction = model.predict(X_test)
```

```python
#examine model parameters 
print('Model Summary:\n')

# Print intercept (alpha)
print('Intercept:')
print('alpha = ' , model.intercept_)

# Print weights
print('\nWeights:')
i = 0
for w in model.coef_:
    print('w',i+1,'= ', w, ' [ weight of ', features[i],']')
    i += 1
```

From these results, Life (healthy life expectancy at birth) has the largest weight so an increase in Life corresponds to an increase in LogGPD.

```python
#evaluate model
#mean squared error
print('\nModel Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, prediction)))
# the coefficient of determination: 1 is perfect prediction
print(' R^2 =   %.2f'
      % r2_score(y_test, prediction))
```

This RMSE score shows that the model can relatively predict the data accurately. The $R^2$ value of 0.80 implies that 80% of the variance in the LogGDP label is explained by the model and the model has a strong fit.

```python
#L1 regularization  

#create lasso model
lasso = Lasso()

# create grid of alpha values 
alpha_values = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}

# grid search to find the best alpha
grid_search = GridSearchCV(lasso, param_grid=alpha_values, cv=10, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# get best alpha 
best_alpha = grid_search.best_params_['alpha']
print(f'Best alpha: {best_alpha}')

# train model with best alpha
lasso_best = Lasso(alpha=best_alpha)
lasso_best.fit(X_train, y_train)
# make predictions on  test data
y_pred_best = lasso_best.predict(X_test)
```

```python
#evaluate model

#mean squared error
print('\nModel Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, y_pred_best)))
# the coefficient of determination: 1 is perfect prediction
print(' R^2 =   %.2f'
      % r2_score(y_test, y_pred_best))
```

```python
# print the coefficients of the features
coefficients = pd.Series(lasso_best.coef_, index=X.columns)

print('Lasso Coefficients:')
print(coefficients)

#features selected by Lasso (non-zero coefficients)
selected_features = coefficients[coefficients != 0].index.tolist()
print('Selected Features:', selected_features)
```

The RMSE and $R^2$ score are the same for both the multiple linear regression model and the lasso regression model. This implies that all the features chosen in my model are relevant and contribute to the prediction.
